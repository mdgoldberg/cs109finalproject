```python
def isPassSM(df):
    if df["isPass"]==True and df["passLoc"] == "SM":
        num = 1
    else:
        num = 0
    return num

def isPassSL(df):
    if df["isPass"]==True and df["passLoc"] == "SL":
        num = 1
    else:
        num = 0
    return num

def isPassSR(df):
    if df["isPass"]==True and df["passLoc"] == "SR":
        num = 1
    else:
        num = 0
    return num

def isPassDM(df):
    if df["isPass"]==True and df["passLoc"] == "DM":
        num = 1
    else:
        num = 0
    return num

def isPassDR(df):
    if df["isPass"]==True and df["passLoc"] == "DR":
        num = 1
    else:
        num = 0
    return num

def isPassDL(df):
    if df["isPass"]==True and df["passLoc"] == "DL":
        num = 1
    else:
        num = 0
    return num

def isRunLE(df):
    if df["isRun"]==True and df["rushDir"] == "LE":
        num = 1
    else:
        num = 0
    return num

def isRunRE(df):
    if df["isRun"]==True and df["rushDir"] == "RE":
        num = 1
    else:
        num = 0
    return num

def isRunM(df):
    if df["isRun"]==True and df["rushDir"] == "M":
        num = 1
    else:
        num = 0
    return num

def isRunLT(df):
    if df["isRun"]==True and df["rushDir"] == "LT":
        num = 1
    else:
        num = 0
    return num

def isRunRT(df):
    if df["isRun"]==True and df["rushDir"] == "RT":
        num = 1
    else:
        num = 0
    return num

def isRunRG(df):
    if df["isRun"]==True and df["rushDir"] == "RG":
        num = 1
    else:
        num = 0
    return num

def isRunLG(df):
    if df["isRun"]==True and df["rushDir"] == "LG":
        num = 1
    else:
        num = 0
    return num

#Create Pass Columns
df["isPassSM"] = df.apply(isPassSM, axis=1)
df["isPassSL"] = df.apply(isPassSL, axis=1)
df["isPassSR"] = df.apply(isPassSR, axis=1)
df["isPassDM"] = df.apply(isPassDM, axis=1)
df["isPassDL"] = df.apply(isPassDL, axis=1)
df["isPassDR"] = df.apply(isPassDR, axis=1)
df["isPassS"] = df["isPassSM"] + df["isPassSL"] + df["isPassSR"]
df["isPassD"] = df["isPassDM"] + df["isPassDL"] + df["isPassDR"]

#Create Run Columns
df["isRunLE"] = df.apply(isRunLE, axis=1)
df["isRunRE"] = df.apply(isRunRE, axis=1)
df["isRunRT"] = df.apply(isRunRT, axis=1)
df["isRunLT"] = df.apply(isRunLT, axis=1)
df["isRunRG"] = df.apply(isRunRG, axis=1)
df["isRunLG"] = df.apply(isRunLG, axis=1)
df["isRunM"] = df.apply(isRunM, axis=1)
df.head()
```

```python
n = float(len(df.passLoc.dropna()) +  len(df["rushDir"].dropna()))
res_dict = {
    "isPassSL": df.isPassSL.sum()/n,
    "isPassSM": df.isPassSM.sum()/n,
    "isPassSR": df.isPassSR.sum()/n,
    "isPassDL": df.isPassDL.sum()/n,
    "isPassDM": df.isPassDM.sum()/n,
    "isPassDR": df.isPassDR.sum()/n,
    "isRunLE": df.isRunLE.sum()/n,
    "isRunLT": df.isRunLT.sum()/n,
    "isRunLG": df.isRunLG.sum()/n,
    "isRunM": df.isRunM.sum()/n,
    "isRunRG": df.isRunRG.sum()/n,
    "isRunRT": df.isRunRT.sum()/n,
    "isRunRE": df.isRunRE.sum()/n
    }
```

```python
def opac_calc(val, upper, minimum, ou, om):
    diff = ou - om 
    diff2 = upper - minimum
    step = diff/diff2
    n = (val-minimum)*step + om
    return n

opac_calc(10.94, 14.76, 2.11, 100.0, 60.0)
```
